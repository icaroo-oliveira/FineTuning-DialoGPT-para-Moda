```python

print(training_sentences[:5])
print(training_intents[:5])
```

```python

intent_labels = ["recomendacao", "tendencia", "busca_loja"]
```

```python
from sentence_transformers import SentenceTransformer
import numpy as np


embedding_model = SentenceTransformer("all-MiniLM-L6-v2")#modelo de embeddings

#converte frases para embeddings
X_train = embedding_model.encode(training_sentences)
y_train = np.array(training_intents)
```

```python
from sklearn.svm import SVC

#treina o classificador
intent_classifier = SVC(kernel="linear", probability=True)
intent_classifier.fit(X_train, y_train)
```

```python
def detect_intent(user_input):
    user_embedding = embedding_model.encode([user_input])
    intent_index = intent_classifier.predict(user_embedding)[0]
    confidence = intent_classifier.predict_proba(user_embedding).max()  #confiança da predição
    return intent_labels[intent_index], confidence

#teste
user_message = "What are some stylish ways to wear leather pieces this season?"
intent_detected, confidence = detect_intent(user_message)
print(f"Intenção detectada: {intent_detected} (Confiança: {confidence:.2f})")
```
